# Make config file

### Option 1:

Make and download base_config.cfg from the spaCy quickstart widget: https://spacy.io/usage/training#quickstart

Fill config file with command:

> python -m spacy init fill-config base_path output_file --diff

Example: 
> python -m spacy init fill-config base_config.cfg config.cfg --diff 

base_config.cfg should be the file you just made and downloaded from the quickstart widget

--diff = prints the differences between base_config and config

### Option 2:

Make and fill config file with one command:
> python -m spacy init config output_file --lang --pipeline --optimize --gpu

Example:
> python -m spacy init config config.cfg -l nl -p ner -o accuracy --gpu

- **-l** = "--lang", we set our language to Dutch (nl)
- **-p** = "--pipeline", we set our pipeline components to only NER
- **-o** = "--optimize", we optimize for accuracy instead of efficiency
- **--gpu** = we enable GPU



More info: https://spacy.io/api/cli#init

# Debug data

In order to debug the data, or check for problems, we first have to manually add our train and dev data paths to the config.cfg file.

This can be done in the very first lines of the file:
> [paths]

> train = path_to_your_data_file/spacy_data_TRAIN.spacy

> dev = path_to_your_data_file/spacy_data_DEV.spacy

Now we can use the command:

> python -m spacy debug data config.cfg

In our case, we have one warning: "Low number of examples to train a new pipeline (124)"

Our data has only 155 charters in total, the training data is only 124.


# Training

Assuming the model will be trained on google colab, you can either replace the paths in the previous step with the corresponding paths in your google drive and use the first command, or leave them blank ("null") and use the second command:

If you filled in the train and dev paths:
> python -m spacy train config_path --output --gpu-id 


- **--output** = directory to output trained models in
- **gpu-id** set to 0 to enable

If you left the paths blank:
> python -m spacy train config.cfg --output your_output_path --paths.train your_train_path --paths.dev your_dev_path --gpu-id 0

Example in Google Colab:
> pip install spacy

> pip install spacy-transformers

> python -m spacy train config.cfg --output drive/MyDrive/spaCy/output --paths.train drive/MyDrive/spaCy/training_data_TRAIN.spacy --paths.dev drive/MyDrive/spaCy/training_data_DEV.spacy --gpu-id 0

### Output interpretation

During training he following information will be displayed:

E = epoch

\# = steps/batches

LOSS TRANSFORMER = loss for transformer pipeline component

LOSS NER = loss for NER pipeline component

ENTS_F = f-score for entities

ENTS_P = precision for entities

ENTS_R = recall for entities

SCORE = overall accuracy

# Package Pipeline



Next, we want to package our pipeline so we can upload it to Huggingface. For this, it was necessary (in my case) to make an Anaconda environment with Python 3.10 and Spacy version 3.5.

Install spacy-huggingface-hub:
> pip install spacy-huggingface-hub

These installations will possibly be necessary too:
> conda install huggingface_hub

> pip install chardet

### IMPORTANT: 
- The model name for Huggingface must be assigned NOW (in the following command)!! If it is not assigned now, changing the name at any point later than this will cause problems.
- The model name must have NO dashes (-), only underscores (_)

> python -m spacy package model_best_path output_dir --build wheel –name pipeline_name 

Example:
> python -m spacy package model-best output --build wheel –name pipeline_name

# Log in to Hugginface

In order to upload our model to Huggingface, we must first log in to Huggingface in the command line:

Enter the following command:
> huggingface-cli login

You will be prompted to input your user authentication token. This can be found in your profile on Huggingface.

You will also be advised to run the following command in case you want to set this credential helper as the default:

> git config --global credential.helper store

# Push to Huggingface

To push your model to Huggingface, use the following command: 

> python -m spacy huggingface-hub push pipeline_name-0.0.0-py3-none-any.whl

pipeline_name will be the name you entered as pipeline name when packaging the pipeline. It seems the language of your model will automatically be added before the name of your pipeline. In our case, the name we used when packaging was "middle_dutch_spacy", it was changed to "nl_middle_dutch_spacy".
