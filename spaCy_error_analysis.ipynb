```python
import spacy
from spacy_error_analysis import spacy_to_doc, print_scores, error_analysis, render_displacy, preds_vs_corr
```

```python
# Define paths to best model and test set

model_path = "drive/MyDrive/DTA/Internship/spaCy/output/model-best"
test_path = "drive/MyDrive/DTA/Internship/spaCy/training_data_TEST.spacy"
```

## Error analysis

```python
# Function that loads model from path, loads test set from path and convert it to doc objects, and prints coorect and predicted labels for entities in specific example number in test set

error_analysis(model_path, test_path, 12)
```

```python
# This does the exact same thing as the previous function, but it's deconstructed. This way you can do the loading and test set conversion only once, and test different examples faster

nlp_ner = spacy.load(model_path)
test_set = spacy_to_doc(test_path)
preds_vs_corr(test_set, 4, nlp_ner)
```

## Scores

```python
# Prints scores (for each entity) of the loaded trained model on the test set

print_scores(test_set, nlp_ner)
```
